**Арифметические** **операторы**


```python
x = 4
y = 3
#+	Addition	
print("+	cложение", x + y	)
#-	Subtraction	
print("-	вычитание", x - y	)

#*	Multiplication	
print("*	умножение", x * y	)
#/	Division	
print("/	деление", x / y	)

#%	Modulus	x % y	
print("%	Деление по модулю (остаток от деления)", x % y	)

#**	Exponentiation	

print("**	Возведение в степень", x ** y	)

#//	Floor division	
print("//	Деление без остатка", x // y	)

```
